```python
# Your code goes here
import pandas as pd

# Load data set
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
df = pd.read_csv(url)

# Step 1: Cleaning Column Names
# Lowercase the column headers, replace spaces with underscores, and replace 'st' with 'state'
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('st', 'state')
# 2.1 Gender column: standardize values to 'male' and 'female'
gender_mapping = {'m': 'male', 'f': 'female', 'female': 'female', 'male': 'male', 'femal': 'female'}
df['gender'] = df['gender'].str.lower().map(gender_mapping)

# Verify gender values
print("Unique values in 'gender' after cleaning:", df['gender'].unique())

# 2.2 State column: replace abbreviations with full names
state_mapping = {
    'california': 'california', 'cali': 'california', 'ca': 'california',
    'arizona': 'arizona', 'az': 'arizona',
    'washington': 'washington', 'wa': 'washington',
    'nevada': 'nevada', 'nv': 'nevada',
    'oregon': 'oregon', 'or': 'oregon'
}
df['state'] = df['state'].str.lower().map(state_mapping)

# Verify state values
print("Unique values in 'state' after cleaning:", df['state'].unique())

# 2.3 Education column: replace 'Bachelors' with 'Bachelor'
df['education'] = df['education'].replace('bachelors', 'bachelor')

# Verify education values
print("Unique values in 'education' after cleaning:", df['education'].unique())


# Print column names to ensure they are correct
print("Column names after cleaning:")
print(df.columns)
```

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here

import pandas as pd

# URL of the dataset
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(url)

# Inspect the DataFrame
df.head()
```

```python
pivot_total_revenue = df.pivot_table(values='total_claim_amount', 
                                     index='sales_channel', 
                                     aggfunc='sum').round(2)
print("Total Revenue by Sales Channel:")
print(pivot_total_revenue)
```

```python
pivot_avg_clv = df.pivot_table(values='customer_lifetime_value', 
                               index=['gender', 'education'], 
                               aggfunc='mean').round(2)

print("Average Customer Lifetime Value by Gender and Education Level:")
print(pivot_avg_clv)
```
